```python
class Jackknife:
    def __init__( self, len_data, binsize ):
        self.binsize = binsize
        self.nbins = math.floor( len_data/self.binsize )
        self.N = self.binsize * self.nbins
        self.jack_avg = []
        self.est = 0
        self.var_est = 0

    def set( self, func, list_of_data ):
        for i in range( self.nbins ):
            self.jack_avg.append( func( i, self.binsize, list_of_data ) )

    def do_it( self ):
        for i in range( 0, self.nbins ):
            self.est += self.jack_avg[i]
        self.est /= self.nbins

        for i in range( 0, self.nbins ):
            self.var_est += ( self.jack_avg[i] - self.est )**2
        self.var_est /= self.nbins
        self.var_est *= self.nbins -1

    def mean( self ):
        return self.est

    def var( self ):
        return self.var_est

    def err( self ):
        return np.sqrt(self.var_est)

def simple_mean(i, binsize, np_data):
    resmpld = np.delete(np_data, np.s_[i*binsize:(i+1)*binsize], axis=0)
    return np.mean(resmpld, axis=0)

def jk_avg(i, binsize, np_data):
    return np_data[i]

def format_print(cen, err):
    for i in range(-50, 50):
        if 10**(-i+1)>=err>10**(-i):
            tmp=err*10**(i+1)
            return '{num:.{width}f}'.format(num=cen, width=i+1)+'('+str(round(tmp))+')'

def format_print_w_exact(exact, cen, err):
    for i in range(-50, 50):
        if 10**(-i+1)>=err>10**(-i):
            tmp=err*10**(i+1)
            return '{num:.{width}f}'.format(num=cen, width=i+1)+'('+str(round(tmp))+')'+': '+'{num:.{width}f}'.format(num=(exact-cen)/err, width=i+1)+' sigma'
```

```python
mass="0p2000"

betas = np.loadtxt( "../spline_cpp/betas_mass"+mass+"_32.dat" )
ibetacs_jk = np.loadtxt( "../spline_cpp/ibetac_jk_mass"+mass+"_"+str(Ns)+".dat" )

betacs_jk = np.array([[betas[ int(ibetacs_jk[jdrop][ibin]) ] for ibin in np.arange(nbins) ] for jdrop in np.arange(nbetas[mass]) ])
Tcs_jk = np.array([[ f_ainv( betacs_jk[jdrop][ibin], 0.2)/Nt for ibin in np.arange(nbins) ] for jdrop in np.arange(nbetas[mass]) ])
MBs_jk = np.array([[ f_MB( betacs_jk[jdrop][ibin], 0.2) for ibin in np.arange(nbins) ] for jdrop in np.arange(nbetas[mass]) ])
```

```python
def Tc_MB_fitter( MB, b0, b1, b2 ):
    return b0 + b1*MB + b2*MB**2
```

```python
def dTc_dMB_fitter( MB, b0, b1, b2 ):
    return b1 + 2.0*b2*MB
```

```python
class F_Tc:
    def fitter( self, MB, b0, b1 ):
        return b0+b1*MB

    def dfitter( self, MB, b0, b1 ):
        return b1

    def __init__( self ):
        self.fp = np.loadtxt( "../tc/coeffs_Tc_MB.dat" )
        self.fcp = np.loadtxt( "../tc/cov_Tc_MB.dat" )
        self.Dfs = [self.Df0, self.Df1]

    def __call__( self, MB ):
        return self.fitter( MB, self.fp[0], self.fp[1] )
    
    def Df0( self, MB ):
        return 1.0
    
    def Df1( self, MB ):
        return MB
    
    def d( self, MB ):
        return self.dfitter( MB, self.fp[0], self.fp[1] )

    def D( self, MB ):
        df = np.array([ f(MB) for f in self.Dfs ])
        return np.sqrt( df@self.fcp@df )
```
